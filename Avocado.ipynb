### Now, we are going to do two things, we want to plot the data for all the uniques regions and compare them, but we have two different type of avocados of each region, so, we will do them separately, first we will look at organic and then we can repeat the process. 

```python
df = pd.read_csv("avocado.csv")
df = df.copy()[df["type"] == "organic"]
df["Date"] = pd.to_datetime(df["Date"])
df.sort_values(by = "Date", ascending = True, inplace = True)


graph_df = pd.DataFrame()

for region in df["region"].unique():
    region_df = df.copy()[df["region"] == region]
    region_df.set_index("Date", inplace = True)
    region_df.sort_index(inplace = True)
    region_df[f"{region}_Price25ma"] = region_df["AveragePrice"].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[f"{region}_Price25ma"]]
        
    else:
        graph_df = graph_df.join(region_df[[f"{region}_Price25ma"]])
        
graph_df.tail(3)
```

# This has now made us a new dataframe with the 25 moving point average price for each region for type avocado and we can now plot it and compare

```python
graph_df.dropna().plot(figsize = (8,5), legend = False)
```

# THIS IS A COMPARISON OF THE 25 MOVING POINT AVERAGE PRICE FOOR EACH UNIQUE REGION FOR TYPE ORGANIC AVOCADO.

```python
df = pd.read_csv("avocado.csv")
df = df.copy()[df["type"] == "conventional"]
df["Date"] = pd.to_datetime(df["Date"])
df.sort_values(by = "Date", ascending = True, inplace = True)


graph_df = pd.DataFrame()

for region in df["region"].unique():
    region_df = df.copy()[df["region"] == region]
    region_df.set_index("Date", inplace = True)
    region_df.sort_index(inplace = True)
    region_df[f"{region}_Price25ma"] = region_df["AveragePrice"].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[f"{region}_Price25ma"]]
        
    else:
        graph_df = graph_df.join(region_df[[f"{region}_Price25ma"]])
        
graph_df.dropna().plot(figsize = (8,5), legend = False)
```
